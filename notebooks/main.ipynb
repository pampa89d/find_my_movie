#### 1. Убираем годы из названий фильмов 

```python
def preproccesing_text(x: pd.Series):
    x = x.str.replace(r'[\s+]', ' ', regex=True)
    x = x.str.replace(r'#\S+', '', regex=True)
    x = x.str.replace(r'http\S+', '', regex=True)
    x = x.str.replace('br ', '')
    x = x.str.replace(r'@\s+[^ ]+|@', '', regex=True)
    x = x.str.replace(r'[^\w\s]', ' ', regex=True)
    x = x.str.replace(r'[^А-Яа-яЁё\s]', '', regex=True)
    x = x.str.replace(r'\b[А-яА-ЯЁё]\b', '', regex=True)
    x = x.str.lower()
    x = x.str.replace(r'\bне\s+(\w+)\b', r'не_\1', regex=True)
    x = x.str.replace(r'\s+', ' ', regex=True).str.strip()

    return x
```

```python
raw_df['title'] = raw_df['title'].str.replace(r'\s\(\d+\)', '', regex=True)
```

```python
raw_df['genres'] = raw_df['genres'].str.replace(r'[\[\]\']', '', regex=True)
```

```python
raw_df['genres'] = raw_df['genres'].str.lower()
raw_df = raw_df[~raw_df['genres'].str.contains('сериалы')]
```

```python
raw_df['genres'] = np.where(raw_df['genres'] == '', 'unknown', raw_df['genres'])
```

```python
raw_df['country'] = raw_df['country'].str.replace(r'[\[\]\']', '', regex=True)
```

```python
raw_df['country'] = np.where(raw_df['country'] == '', 'unknown', raw_df['country'])
```

```python
def parse_duration(duration):
    if pd.isna(duration):
        return None
    hours = re.search(r'(\d+)\s*ч', duration)
    minutes = re.search(r'(\d+)\s*мин', duration)
    h = int(hours.group(1)) if hours else 0
    m = int(minutes.group(1)) if minutes else 0
    return h * 60 + m
```

```python
raw_df['duration'] = raw_df['duration'].apply(parse_duration)
```

```python
raw_df['description'] = preproccesing_text(raw_df['description'])
```

```python
raw_df = raw_df[raw_df['description'] != '']
```

```python
raw_df['rating_stars'] = raw_df['rating_votes'].str.replace(r'/.*', '', regex=True)
raw_df['rating_stars'] = raw_df['rating_stars'].str.replace(r'{rating.value}', '0', regex=True).fillna(0).astype('float32')
raw_df['rating_voices'] = raw_df['rating_votes'].str.findall(r'\((\d+)+').str[0].fillna(0).astype('int32')
```

```python
raw_df['rating_stars']  = (((raw_df['rating_voices'] / (raw_df['rating_voices'] + 50)) * raw_df['rating_stars']) + \
                                                                                    (50 / (raw_df['rating_voices'] + 50)) * (raw_df['rating_stars'].mean() * \
                                                                                                                    (raw_df['rating_voices'] / raw_df['rating_voices']))).fillna(0)

# raw_df['rating_stars'] = raw_df['rating_stars'] * 2
```

```python
raw_df['film_rating_kp'] = raw_df['film_rating_kp'].str.replace(r'KP ', '', regex=True).astype('float32').fillna(0.)
```

```python
raw_df['film_rating_imdb'] = raw_df['film_rating_imdb'].str.replace(r",'isActive':true", '', regex=True)
```

```python
raw_df['film_rating_imdb'] = raw_df['film_rating_imdb'].str.replace(r'IMDB ', '', regex=True).astype('float32').fillna(0.)
```

```python
df = pd.concat([raw_df.iloc[:,9], raw_df.iloc[:,10], raw_df.iloc[:,11]], axis=1)
average = df.replace(0, np.nan).mean(axis=1).fillna(0.)
raw_df['rating_average'] = average
```
